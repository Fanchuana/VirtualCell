```python
def print_spaced(string:str, h):
    print('  '*h + string)

def show_h5_dataset(obj, h):
    # 如果是数据集
    print_spaced(f"形状: {obj.shape}", h)
    print_spaced(f"数据类型: {obj.dtype}", h)
    print_spaced(f"大小: {obj.size}", h)
        
    # 如果数据量不大，可以查看实际数据
    if obj.size < 100:  # 只查看小数据集
        print_spaced(f"数据: {obj[()]}", h)
    else:
        print_spaced(f"数据过大，只显示前10个元素: {obj[:10]}", h)

def show_h5_obj(father_obj, h):
    for name, obj in father_obj.items():
        print_spaced(f"对象: {name}", h)
        print_spaced(f"类型: {type(obj)}", h+1)

        if isinstance(obj, h5py.Dataset):
            # 如果是数据集
            show_h5_dataset(obj, h+1)
        
        elif isinstance(obj, h5py.Group):
            # 如果是子组
            print('  '*(h+1) + f"子组成员: {list(obj.keys())}")
            print_spaced("-" * 30, h+1)
            show_h5_obj(obj, h+1)
        
        print_spaced("-" * 30, h)

def print_h5_part(part):
    print(f"组路径: {part.name}")
    print(f"组成员数量: {len(part)}")
    print("组成员:", list(part.keys()))
    print("=" * 50)
    
    show_h5_obj(part, 0)
```

```python
import h5py
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import scipy.sparse as sp

# 打开文件
data = h5py.File("/work/home/cryoem666/xyf/temp/pycharm/state/data/competition_train.h5", "r")

# 1. 加载稀疏矩阵数据
print("=== 加载稀疏矩阵数据 ===")
X_data = data['X/data'][:] # type: ignore
X_indices = data['X/indices'][:]
X_indptr = data['X/indptr'][:]
shape = (len(X_indptr)-1, 18080)  # 221273 cells × 18080 genes

# 创建CSR稀疏矩阵
X_sparse = sp.csr_matrix((X_data, X_indices, X_indptr), shape=shape)
print(f"表达矩阵形状: {X_sparse.shape}")
print(f"稀疏矩阵密度: {X_sparse.nnz / (X_sparse.shape[0] * X_sparse.shape[1]):.4f}")

# 2. 创建AnnData对象
gene_names = data['var/_index'][:].astype(str)
cell_names = data['obs/_index'][:].astype(str)

adata = sc.AnnData(X=X_sparse)
adata.var_names = gene_names
adata.obs_names = cell_names

# 3. 添加观察数据 (obs)
print("\n=== 添加元数据 ===")
# 处理分类变量
categorical_vars = ['batch', 'batch_var', 'cell_type', 'guide_id', 'target_gene']
for var in categorical_vars:
    if var in data['obs']:
        categories = data[f'obs/{var}/categories'][:].astype(str)
        codes = data[f'obs/{var}/codes'][:]
        adata.obs[var] = pd.Categorical.from_codes(codes, categories)

print("元数据列:", adata.obs.columns.tolist())
```

```python
# 4. 数据概览
print("\n=== 数据概览 ===")
print(f"细胞数: {adata.n_obs:,}")
print(f"基因数: {adata.n_vars:,}")
print(f"总非零值: {adata.X.nnz:,}")

print(f"\n细胞类型分布:")
print(adata.obs['cell_type'].value_counts())

print(f"\n批次分布 (前10):")
print(adata.obs['batch'].value_counts().head(10))

print(f"\n靶基因分布 (前10):")
print(adata.obs['target_gene'].value_counts().head(10))

print(f"\nguide_id分布 (前10):")
print(adata.obs['guide_id'].value_counts().head(10))
```

```python
# 5. 质量控制
print("\n=== 质量控制 ===")
# 计算QC指标
sc.pp.calculate_qc_metrics(adata, percent_top=None, log1p=False, inplace=True)

# 识别线粒体基因
mt_genes = [gene for gene in adata.var_names if gene.startswith('MT-')]
if len(mt_genes) > 0:
    # 计算线粒体基因百分比
    mt_counts = np.array(adata[:, mt_genes].X.sum(axis=1)).flatten()
    total_counts = np.array(adata.X.sum(axis=1)).flatten()
    adata.obs['percent_mt'] = mt_counts / total_counts * 100
else:
    adata.obs['percent_mt'] = 0

print("QC统计摘要:")
print(f"每个细胞平均计数: {adata.obs['total_counts'].mean():.0f}")
print(f"每个细胞平均基因数: {adata.obs['n_genes_by_counts'].mean():.0f}")
print(f"线粒体百分比: {adata.obs['percent_mt'].mean():.2f}%")
```
