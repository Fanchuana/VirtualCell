## 老训练数据的HVG

```python
import h5py
import pandas as pd

def byte2str(x):
    return x.decode("utf-8") if isinstance(x, (bytes, bytearray)) else str(x)

h5ad_path = "/work/home/cryoem666/xyf/temp/pycharm/state/gene_perturnb_state/data/replogle.h5ad"
data = h5py.File(h5ad_path, "r")

hvg_mask = data["/var/highly_variable"][:]
all_gene_names = data["var/gene_name_index"][:]
hvg_names_raw = all_gene_names[hvg_mask]
hvg_names = [byte2str(name) for name in hvg_names_raw]

print("==== replogle.h5ad 原数据 ====")
print(all_gene_names.shape)
print(all_gene_names)
print(len(hvg_names))
print(hvg_names)
```

## 比对新老数据的HVG

```python
adata = ad.read_h5ad(data_path)

print("==== adata.var 预览 ====")
print(adata.var.head())

cell_line_list = ['jurkat', 'hepg2', 'k562', 'rpe1']

for cell_line in cell_line_list:
    print("="*10 + f"{cell_line}" + "="*10)
    p_vals = adata.varm[f"{cell_line}_" + 'DE_pval'].T
    logfcs = adata.varm[f"{cell_line}_" + 'DE_log2FC'].T

    print("====p_vals====")
    print(p_vals)
    # print(p_vals.loc[['NOC2L']])

    print("====logfcs====")
    print(logfcs)

    existing_hvg_names = p_vals.columns.intersection(hvg_names)
    print('====existing_hvg_names====')
    print(len(existing_hvg_names))
    print(existing_hvg_names)
```
